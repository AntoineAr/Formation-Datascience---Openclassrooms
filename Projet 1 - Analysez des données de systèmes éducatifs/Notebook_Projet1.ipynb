On crée une fonction qui donnera pour chaque DF une description sommaire :

```python
def description_df(df):
    
    # On affiche les dimensions (nb lignes et colonnes) du DF :
    df.shape
    print("Les dimensions du data frame sont de :", df.shape[0], 
          "lignes, et :", df.shape[1], "colonnes.\n")
    print('-' * 50, '\n')
    
    # On affiche les 5 premières lignes du DF :
    print("Visualisation des 5 premières lignes du data frame :")
    display(df.head())
    print('-' * 50)
    
    # On affiche le nombre total d'éléments du DF :
    print("Le data frame comprend", df.size, "éléments.\n")
    
    # On affiche le type de données contenues dans le DF :
    print('-' * 50, "\n")
    print(df.dtypes.value_counts())
    print('-' * 50, "\n")
    
    # On calcule et affiche le nombre total de valeurs manquantes du DF :
    df_na = df.isna().sum().sum()
    print("Il y a", df_na, "valeurs manquantes dans ce data frame.\n")
    
    # Calcul et affichage de la proportion de valeurs manquantes du DF :
    df_prop_na = round((df_na / df.size) * 100, 2)
    print("La proportion de valeurs manquantes y est donc de :", df_prop_na, "%.\n")
    print('-' * 50, "\n")
    
    # Représentation graphique des nombres d'éléments et de valeurs manquantes :
    plt.figure()
    plt.bar(height = 100, x = 'Nombre de Données total (%)')
    plt.bar(height = df_prop_na, x = 'Proportion de données manquantes (%)')
```

Utilisation de la fonction 'description_df' sur chaque data frame :

```python
# Sur le DF principal : 

description_df(data)
```

```python
# Sur le DF 'Series' :
description_df(data_series)
```

```python
# Sur le DF 'Country' : 

description_df(data_country)
```

```python
# Sur le DF 'Country-Series'

description_df(data_cs)
```

### 2.1 Visualisation et analyse des données manquantes

Regardons maintenant le taux de valeurs manquantes par colonne, pour l'ensemble des variables :

```python
print(data.loc[:, '2010':].isna().mean())
msno.matrix(data.loc[:, '2010':])
```

Nous constatons que pour les variables sélectionnées, nous ne disposons d'aucune donnée pour les années au-delà de 2016. Notre analyse se portera sur les dernière valeurs disponibles, en particulier celles de 2015 et 2016. Nous supprimons donc les colonnes de 2017 à 2025.

```python
data = data.dropna(axis = 1, how = 'all')
```

Nous pouvons également regarder le nombre de valeurs manquantes pour chacune des variables que nous avons sélectionnées. Pour cela nous allons créer deux fonctions que nous appelerons pour chaque indicateur.

```python
def na_indicateur(indicateur):
    print("Observation du pourcentage de valeur manquante par année, pour la variable :", indicateur,":")
    display((data.loc[data['Indicator Name'] == indicateur, "2010":].isna().mean()) * 100)

def bar_na(indicateur):
    msno.bar(data.loc[data['Indicator Name'] == indicateur, '2010':])
    plt.show()
```

Utilisons maintenant une boucle 'for' appelant cette fonction pour chacune de nos variables : (à voir)

```python
# Au départ j'avais fait cette boucle for, j'espèrais que pour chaque indicateur, cela m'afficherait le %age
# de valeurs manquantes, puis le graphique msno.bar ensuite. Mais je ne sais pas pourquoi ça n'affiche que les %ages
# et le graphique pour le dernier indicateur seulement.

for i in liste_var:
    na_indicateur(i)
    bar_na(i)

#Je les appelle donc individuellement :
```

Commentaires :

Pour la population totale : on observe un faible nombre de données manquantes pour la période analysée.

Pour la population des 15-24 ans : on voit que cette donnée est totalement manquante pour l'année 2016. Pour les années précédentes, le pourcentage de valeurs manquantes est strictement identique, on peut donc supposer que ce sont les mêmes pays pour laquelle la donnée n'est jamais renseignée.

Pour le taux de croissance de la population (%) : on retrouve les même taux que pour la population totale.

Pour le taux de scolarisation dans le secondaire : le nombre de données manquantes est plus élevé, il est proche des 100% pour l'année 2016. Nous réaliserons des imputations car c'est une variable importante pour notre analyse. Nous sélectionnerons l'année 2015.

Pour le taux de scolarisation dans le tertiaire : là encore on note un fort taux de données manquantes, c'est malgré tout un indicateur important pour notre analyse et, comme pour la variable précédente, nous réaliserons des imputations.

Pour le taux d'achèvement dans le secondaire : même constat ici que pour la variable précédente.

Pour le nombre d'ordinateur personnel (pour cent personnes) : sur la période sélectionnée, cette donnée n'est jamais renseignée, nous la supprimerons donc du dataframe.

Pour le nombre d'utilisateurs d'Internet (pour cent personnes) : donnée plutôt bien renseignée sur notre période d'analyse. C'est un bon point car elle nous servira d'indicateur technologique pour chaque pays.

Pour le PIB/tête : donnée plutôt bien reseignée, même si un pourcentage de valeurs manquantes autour de 15% n'est pas négligeable.

Pour les dépenses publiques en éducation en pourcentage du PIB : on a ici une part de données manquantes plutôt élevée. Nous verrons ensuite si la variable peut ou nous être conservée dans le data frame.

Pour la part d'investissement de long terme dans l'ensemble des dépenses dans les institutions publiques : ici nous considérons que la donnée est trop mal renseignée et nous devrons la supprimer du data frame.

Pour la part des dépenses courantes dans l'ensemble des dépenses concernant les institutions publiques : même chose que pour la variable précédentes, le nombre de données manquantes est trop important pour conserver la variable.

Nous créons une fonction retournant des graphiques permettant la visualisation de la dispersion de chaque variable :

```python
def graphdisp(df, indicateur):
    fig = plt.figure(figsize=(10,8))
    fig.suptitle(f"Distribution de la variable {indicateur}")
    ax0 = plt.subplot2grid((2, 2), (0,0), colspan = 1)
    ax0.hist(df[indicateur])
    ax0.set(xlabel = indicateur)
    ax0.set(ylabel = "Occurences")
    ax0.set_title('Histogramme')

    ax1 = plt.subplot2grid((2, 2), (0,1), colspan = 1)
    ax1.boxplot(df[indicateur])
    ax1.set(ylabel = indicateur)
    ax1.set_title('Boîte à moustaches')
    
    ax2 = plt.subplot2grid((2, 2), (1,0), colspan = 2)
    sns.boxplot(df, x = df[indicateur], y = data_pays['Region'])
    ax2.set(ylabel = None)
    
    fig.tight_layout(pad=2.0)
    plt.show()
```

Nous créons une fonction retournant les statistiques descriptives de chaque variable :

```python
# Création d'une fonction retournant les statistiques descriptives "de base" de chaque variable :

def statbase(df, indicateur):
    print(f"Statistiques descriptives de la variable {indicateur}")
    display(df[indicateur].describe())
    mean_var = round(df[indicateur].mean(), 2)
    std_var = round(df[indicateur].std(), 2)
    med_var = round(df[indicateur].median(), 2)
    Q1 = np.percentile(df[indicateur], 25, method = 'midpoint')
    Q3 = np.percentile(df[indicateur], 75, method = 'midpoint')
    IQR = round((Q3 - Q1), 2)
    print(f"La moyenne de cette variable est de {mean_var}.\n")
    print(f"Sa médiane est de {med_var}.\n")
    print(f"Son écart-type de : {std_var}.\n")
    print(f"L'écart interquartile de cette variable est de {IQR}.")
```

### 4.1 Construction des indicateurs et scores "intermédiaires"

On crée une variable synthétique pour les taux de scolarisation secondaire et tertiaire, on fait en réalité la moyenne des deux indicateurs, mais le public ciblé par Académy étant les lycéens et les gens entreprenant des études supérieures, nous décidons de donner un peu plus de poids aux taux de scolarisation tertiaire.

```python
# Création d'une variable 'Scolarisation' qui équivaut à la moyenne des taux de scolarisation dans le 2ndaire et le tertiaire
# data_pays['Scolarisation'] = (data_pays['Tx scolarisation secondaire (%)'] + data_pays['Tx scolarisation tertiaire (%)']) / 2

# En donnant plus de poids au tertiaire :
data_pays['Scolarisation'] = ((0.8 * data_pays['Tx scolarisation secondaire (%)']) + (1.2 * data_pays['Tx scolarisation tertiaire (%)'])) / 2
```

On créé une variable 'Population concernée' qui correpond au ratio de la population des 15-24 ans sur la population totale, de manière à "corriger" un certain effet taille des pays.

```python
# Création d'une variable 'Population concernée' qui équivaut au ratio Population 15-24 / Population totale x 100 :
data_pays['Population_concernée'] = (data_pays['Population des 15-24 ans'] / data_pays['Population totale']) * 100
```

Nous allons désormais créer 4 scores :
   - démographique;
   - économique;
   - technologique;
   - scolaire;
   
  
Ils seront tous construits de la même façon à savoir que nous choisissons de faire le ratio entre l'indicateur d'un pays et la valeur maximale de cet indicateur. Ainsi tous ces scores seront compris entre 0 et 1.

Concernant le score démographique, nous hésitons entre l'utilisation de la nouvelle variable 'Population concernée' ou utiliser la variable "brute" de la population des 15-24 ans.
La première option est choisie ici.

```python
# Création du score 'population'
data_pays['SCORE_POP'] = data_pays['Population_concernée'] / data_pays['Population_concernée'].max()

# Je tente en utilisant ici la pop des 15-24 ans :
# data_pays['SCORE_POP'] = data_pays['Population des 15-24 ans'] / data_pays['Population des 15-24 ans'].max()
```

```python
# Création du score 'scolarisation'
data_pays['SCORE_SCO'] = data_pays['Scolarisation'] / data_pays['Scolarisation'].max()
```

```python
# Création du score 'technologie'
data_pays['SCORE_TECHNO'] = data_pays["Utilisateurs d'Internet (pour 100 personnes)"] / data_pays["Utilisateurs d'Internet (pour 100 personnes)"].max()
```

```python
# Création du score 'économie' :
data_pays['SCORE_ECO'] = data_pays['PIB/tête, PPP ($ contant 2011)'] / data_pays['PIB/tête, PPP ($ contant 2011)'].max()
```

Je crée une fonction de score synthétique, avec coefficients ajustables, de manière à pouvoir pondérer de diverses façons en fonction des indicateurs.

```python
coef_sco = coef_techno = coef_eco = coef_pop = 0
data_pays_sco['SCORE_SYNT'] = 0
data_pays_scograph = 0

def scoring(coef_sco = 0.25, coef_techno = 0.25, coef_eco = 0.25, coef_pop = 0.25):
    data_pays_sco['SCORE_SYNT'] = (coef_sco * data_pays_sco['SCORE_SCO']) + (coef_techno * data_pays_sco['SCORE_TECHNO']) + (coef_eco * data_pays_sco['SCORE_ECO']) + (coef_pop * data_pays_sco['SCORE_POP'])
    data_pays_scograph = data_pays_sco.sort_values('SCORE_SYNT', ascending = False).head(15)
    display(data_pays_scograph)
    print("-"*120)
    plt.figure(figsize = (10,6))
    # plt.bar(height = data_pays_scograph['SCORE_SYNT'], x = data_pays_scograph['Country Name'], label = 'SCORE')
    sns.barplot(data = data_pays_scograph, x = 'Country Name', y = "SCORE_SYNT", hue = "Region", dodge = False)
    plt.xticks(rotation='vertical')
    plt.ylabel("Score final")
    plt.legend(bbox_to_anchor=(1, 1.02))
    plt.grid(axis = 'y', linewidth=0.5)
    plt.title("Les 15 pays ayant le meilleur potentiel pour un développement d'Academy à l'international")
    plt.show()
```

Par défaut on a indiqué une pondération équivalente pour chaque score intermédiaire : 'Scolarisation', 'Technologie', 'Economie' et 'Population'.

**Score pays en utilisant une pondération égale pour chaque score intermédiaire :**

```python
scoring()
```

Etant donnée notre problématique, il peut sembler raisonnable d'accorder plus de poids aux scores 'Scolarisation' et 'Technologie', et peut-être un peu moins pour les score 'Economie' et 'Population'. 
Voyons les résultats avec les valeurs respectives de : 0.35, 0.35, 0.15 et 0.15 pour les différents coefficients.

**Score par pays en privilégiant les indicateurs de scolarisation et de technologie :**
